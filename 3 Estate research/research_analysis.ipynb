Основная цель исследования - выявить интересные особенности и зависимости, которые существуют на рынке недвижимости, установить параметры, определяющие рыночную стоимость объекта недвижимости

- Добавление вспомогательных столбцов

### Предобработка данных

#### Заполнение пропусков

Первым делом необходимо заполнить выявленные пропуски.
Полезно изучить перечень уникальных значений для каждого столбца

```python
for x in df.columns:
    display(x, df[x].sort_values().unique())
```

Выведем долю пропусков для каждого поля

```python
for x in df.columns:
    print('Доля пропусков в поле', x, '-', round(df[x].isna().sum() / len(df) * 100, 2), '%')
```

Особое внимание нужно уделить полям 'ceiling_height', 'is_apartment', 'balcony', 'airports_nearest', 'cityCenters_nearest', 'parks_around3000', 'parks_nearest', 'ponds_around3000', 'ponds_nearest' - там пропусков больше всего.

Займемся заполнением пропусков. В первых 5-ти столбцах пропусков нет. 6-й столбец - высота потолка, и там много пропущенных данных (скорее всего продавец не указал информацию). Можно попробовать заполнить их медианными значениями по группам квартир из домов с разной этажностью.

```python
for i in df['floors_total'].unique():
    df.loc[(df['floors_total'] == i) & (df['ceiling_height'].isna()), 'ceiling_height'] = \
    df.loc[(df['floors_total'] == i) & (~df['ceiling_height'].isna()), 'ceiling_height'].median()
df.info()
```

Пропуски все равно остались, но в очень малом количестве. Можно удалить.

```python
df = df.dropna(subset=['ceiling_height']).reset_index(drop=True)
df.info()
```

В поле с общим числом этажей есть пропуски (видимо, так же не указали информацию). Заполнить эти поля не получится. Посмотрим, сколько пропусков:

```python
df['floors_total'].isna().sum() / len(df)
```

Пропусков мало, всего 0,038 %. Можно удалить без существенных потерь.

```python
df = df.dropna(subset=['floors_total']).reset_index(drop=True)
df.info()
```

Пропуски в поле с жилой площадью можно заполнить медианными данными по группам квартир, сгруппированным по их общей площади (в этих данных пропусков нет)
Сразу же по такой же схеме можно заполнить и пропуски в поле с площадью кухонь.

```python
for i in df['total_area'].unique():
    df.loc[(df['total_area'] == i) & (df['living_area'].isna()), 'living_area'] = \
    df.loc[(df['total_area'] == i) & (~df['living_area'].isna()), 'living_area'].median()
    df.loc[(df['total_area'] == i) & (df['kitchen_area'].isna()), 'kitchen_area'] = \
    df.loc[(df['total_area'] == i) & (~df['kitchen_area'].isna()), 'kitchen_area'].median()
df.info()
```

Видно, что пропуски в значениях все равно остались, но в гораздо меньшем количестве.
Посчитаем, сколько их в процентном соотношении.

```python
display(df['living_area'].isna().sum() / len(df), df['kitchen_area'].isna().sum() / len(df))
```

И там, и там - меньше 1%. Можно удалять.

```python
df = df.dropna(subset=['living_area', 'kitchen_area']).reset_index(drop=True)
df.info()
```

Далее, имеются пропуски в поле 'is_apartment'. По всей видимости, при подаче объявления, это поле просто не было заполнено. Значит, есть смысл заполнить пропуски значением False.

```python
df['is_apartment'] = df['is_apartment'].fillna('False')
df.info()
```

Далее, поле 'balcony'. Тут пропущенные значения заменим нулем.

```python
df['balcony'] = df['balcony'].fillna(0)
df.info()
```

Поле 'locality_name' не заполнить никак. Посмотрим, сколько в нем пустых значений. Если мало - можно удалить.

```python
df['locality_name'].isna().sum() / len(df)
```

```python
df = df.dropna(subset=['locality_name']).reset_index(drop=True)
df.info()
```

Отсутствующие значения в полях о близости к аэропорту и центру города заполнить так же не получится. Видимо, это значит, что квартира находится на очень большом удалении и от того, и от другого. Пропусков много, поэтому придется их оставить.

В полях с количеством близких парков и прудов можно заменить пропуски на ноль - если значение не указали, логично, что их поблизости нет. Пропуски в расстояниях до парков и прудов придется оставить.

```python
df['parks_around3000'] = df['parks_around3000'].fillna(0)
df['ponds_around3000'] = df['ponds_around3000'].fillna(0)
df.info()
```

Осталось поле с количеством дней с даты публикации до продажи квартиры. Тут пропуски, по всей видимости, обусловлены тем, что квартира до сих пор не продана. Заполнить их не получится. Таких строк много, больше 3000, так что лучше их оставить.

Итак, пропуски заполнены, кроме 5-ти полей. Пришлось удалить 410 строк, что является незначительным количеством.

#### Изменение типов данных

Далее, следует изменить типы данных в ряде полей. 

```python
df.info()
```

Первым делом, для удобства, целесообразно привести столбцы floors_total, balcony, parks_around3000, ponds_around3000 к целочисленному типу. Данные в этих столбцах не могут быть дробными, так что смысла оставлять их в вещественном типе нет.	

```python
df['floors_total'] = df['floors_total'].astype('int')
df['balcony'] = df['balcony'].astype('int')
df['parks_around3000'] = df['parks_around3000'].astype('int')
df['ponds_around3000'] = df['ponds_around3000'].astype('int')
df.info()
```

Полю 'first_day_exposition' есть смысл задать тип данных "datetime".

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
df.info()
```

Теперь остается изменить тип данных в поле 'is_apartment' на логический.

```python
df['is_apartment'] = df['is_apartment'].astype('bool')
df.info()
```

Для экономии памяти, заменим тип данных с float64 на float32 и с int64 на int32.

```python
for x in df.columns:
    if df[x].dtypes == 'int64':
        df[x] = df[x].astype('int32')
    elif df[x].dtypes == 'float64':
        df[x] = df[x].astype('float32')
df.info()
```

#### Удаление неявных дубликатов

Далее, нужно выявить неявные дубликаты в поле 'locality_name'. Рассмотрим, какие значения имеются.

```python
df['locality_name'].sort_values().unique()
```

Видно 2 проблемы:

- нет единообразия в использовании букв "е" и "ё" (к примеру, "поселок" и "посёлок")
- возможна путаница с типом населенного пункта (к примеру, есть "городской поселок", есть "поселок городского типа", а есть населенные пункты вообще буз упоминания типа). Удобнее оставить только названия населенных пунктов, без их типов. Тогда и алфавитную сортировку можно будет сделать проще.

```python
df['locality_name'] = pd.Series(df['locality_name']).str.replace('ё', 'е', regex=True)
df['locality_name'].sort_values().unique()
```

Первая проблем устранена. Теперь можно избавиться от дубликатов - убрать из столбца 'locality_name' упоминание типа населенного пункта, оставив только название.

```python
local_list = ['городской поселок ', 'деревня ', 'коттеджный поселок ', 'поселок городского типа ', \
'поселок при железнодорожной станции ', 'поселок станции ', 'садоводческое некоммерческое товарищество ', \
'садовое товарищество ', 'поселок ', 'село ']
for c in local_list:
    df['locality_name'] = df['locality_name'].str.replace(c, '')
```

```python
df['locality_name'].sort_values().unique()
```

Стоимость квартры напрямую зависит от общей площади и от количества комнат. Наиболее дешевые квартиры на первом и последнем этаже. Также, в последние 2 года имеется тенденция к увеличению стоимости квартир.

#### Зависимость стоимости квартиры от расстояния до центра города

Цена объектов равномерно снижается с удалением от центра. Однако имеются пики на отметках 35, 45 и 55 км, причем на 55 км стоимость практически такая же, как в центре, после чего следует резкий провал. Такое большое расстояние может указывать, что объект находится, к примеру, в элитном пригороде.
