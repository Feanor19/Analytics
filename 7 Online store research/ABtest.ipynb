Пропусков в данных нет, типы данных соответствуют. Оценки каждого из параметров корректно распределены в диапазоне от 1 до 10.

### Приоритизация гипотез фреймворком ICE

Для приоритизации методом ICE (от англ. impact, confidence, effort - «влияние, уверенность, усилия »), воспользуемся формулой: ICE score = Impact * Confidence / Efforts.

Добавим в датафрейм соответствующее поле, а затем отсортируем данные по нему.

```python
df['ICE'] = round(df['Impact'] * df['Confidence'] / df['Efforts'], 3)
df.sort_values(by='ICE', ascending=False)
```

Из результирующего датафрейма видно, что наибольшую оценку получают гипотезы:
- № 8 (Запустить акцию, дающую скидку на товар в день рождения), 
- № 0 (Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей), 
- № 7 (Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок).

Худшая оценка - у гипотез:
- № 4 (Изменить цвет фона главной страницы, чтобы увеличить вовлеченность пользователей), 
- № 3 (Изменить структура категорий, что увеличит конверсию, т.к. пользователи быстрее найдут нужный товар),
- № 5 (Добавить страницу отзывов клиентов о магазине, что позволит увеличить количество заказов).

### Приоритизация гипотез фреймворком RICE

Для приоритизации методом RICE (R от англ. reach, «охват»), воспользуемся формулой: RICE score = Reach * Impact * Confidence / Efforts.

Добавим в датафрейм соответствующее поле, а затем отсортируем данные по нему.

```python
df['RICE'] = df['Reach'] * df['Impact'] * df['Confidence'] / df['Efforts']
df.sort_values(by='RICE', ascending=False)
```

Из результирующего датафрейма видно, что наибольшую оценку получают гипотезы:
- № 7 (Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок), 
- № 2 (Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа), 
- № 0 (Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей),
- № 6 (Показать на главной странице баннеры с актуальными акциями и распродажами, чтобы увеличить конверсию).

Худшая оценка - у гипотез:
- № 4 (Изменить цвет фона главной страницы, чтобы увеличить вовлеченность пользователей), 
- № 5 (Добавить страницу отзывов клиентов о магазине, что позволит увеличить количество заказов),
- № 1 (Запустить собственную службу доставки, что сократит срок доставки заказов).

### Промежуточный вывод

При применении фреймворка RICE, получили приоритизацию гипотез, несколько отличную от ICE. Так, в частности, первое место заняла гипотеза № 7 (3-е место по оценке ICE), а на втором месте гипотеза № 2, которая при приоритизации ICE демонстрировала средние значения. Аналогичные изменения можно проследить и среди "аутсайдеров".

Причиной указанных изменений является использование в расчете RICE параметра Reach (охват). У гипотезы № 7 - максимальный охват (10), поэтому она лидидирует с отрывом. У гипотезы № 2 тоже большой охват (8), поэтому при средних значениях прочих параметров она занимает уверенное второе место.

### График кумулятивной выручки по группам

Вычислим кумулятивные данные по каждому из двух датафреймов и объединим их в датафрейм cumulativeData

```python
datesGroups = orders[['date', 'group']].drop_duplicates()
```

```python
ordersAggregated = datesGroups.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], \
orders['group'] == x['group'])].agg({'date' : 'max', 'group' : 'max', 'transaction_id' : 'nunique', \
'visitor_id' : 'nunique', 'revenue' : 'sum'}), axis=1).sort_values(by=['date','group'])
```

```python
visitorsAggregated = datesGroups.apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], \
visitors['group'] == x['group'])].agg({'date' : 'max', 'group' : 'max', 'visitors' : 'sum'}), \
axis=1).sort_values(by=['date','group'])
```

```python
cumulativeData = ordersAggregated.merge(visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
cumulativeData.head()
```

Разделим общий датафрейм на два отдельных по каждой группе.

```python
cumulativeRevenueA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders']]
cumulativeRevenueB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders']]
```

Построим график кумулятивной выручки по группам.

```python
fig, ax = plt.subplots(figsize = (12,5))
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], label='B')
plt.legend()
plt.title('График кумулятивной выручки по группам')
plt.xlabel('Дата')
plt.ylabel('Кумулятивная выручка');
```

Выручка по группе A увеличивается почти равномерно на протяжении всего периода.

Выручка по группе B увеличивается почти равномерно (чуть лучше группы A) с 01.08 по 18.08, а затем делает резкий скачок. Наблюдается явная аномалия - либо всплеск количества заказов, либо очень дорогой заказ. После 19.08 - также равномерный рост


### График кумулятивного среднего чека по группам

```python
fig, ax = plt.subplots(figsize = (12,5))
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue']/cumulativeRevenueA['orders'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue']/cumulativeRevenueB['orders'], label='B')
plt.legend() 
plt.title('График кумулятивного среднего чека по группам')
plt.xlabel('Дата')
plt.ylabel('Кумулятивный средний чек');
```

По группе A средний чек установился на стабильном уровне - в районе отметки 7000. Прослеживается небольшая тенденция к росту.

В группе B, по всей видимости, 18.08 был очень дорогой заказ - это объясняет резкий взлет кумулятивного среднего чека 18.08 и дальнейшее его падение. По этой группе средний чек еще не установился, но имеется тенденция к снижению.

### График  относительного изменения кумулятивного среднего чека группы B к группе A

```python
mergedCumulativeRevenue = cumulativeRevenueA.merge(cumulativeRevenueB, left_on='date', right_on='date', \
how='left', suffixes=['A', 'B'])
```

```python
fig, ax = plt.subplots(figsize = (12,5))
plt.plot(mergedCumulativeRevenue['date'], (mergedCumulativeRevenue['revenueB']/mergedCumulativeRevenue['ordersB'])/(mergedCumulativeRevenue['revenueA']/mergedCumulativeRevenue['ordersA'])-1)
plt.axhline(y=0, color='black', linestyle='--') 
plt.title('График относительного изменения кумулятивного среднего чека группы B к группе A')
plt.xlabel('Дата')
plt.ylabel('Относительное изменение кумулятивного среднего чека');
```

Для группы B значение кумулятивного среднего чека лучше почти на всем периоде тестирования (за исключением нескольких дней в середине месяца. Однако причиной такого преимущества может быть скачок 18.08. Не будь его, значения по двум группам могли бы отличаться не существенно.

### Статистическая значимость различий в среднем количестве заказов между группами по «сырым» данным

Первым делом сформулируем гипотезы. 

**Нулевая**: различий в среднем количестве заказов между группами нет. 

**Альтернативная**: различия в среднем количестве заказов между группами есть.

В качестве подготовки к анализу, сформируем сводную таблицу data.
В таблицу войдут данные, сгруппированные по дате и по группе - количество заказов, общая сумма заказов и количество пользователей, а также кумулятивные данные по указанным параметрам.

```python
visitorsADaily = visitors[visitors['group'] == 'A'][['date', 'visitors']]

visitorsADaily.columns = ['date', 'visitorsPerDateA']

visitorsACummulative = visitorsADaily.apply(lambda x: visitorsADaily[visitorsADaily['date'] \
<= x['date']].agg({'date': 'max', 'visitorsPerDateA': 'sum'}), axis=1)

visitorsACummulative.columns = ['date', 'visitorsCummulativeA']

visitorsBDaily = visitors[visitors['group'] == 'B'][['date', 'visitors']]

visitorsBDaily.columns = ['date', 'visitorsPerDateB']

visitorsBCummulative = visitorsBDaily.apply(lambda x: visitorsBDaily[visitorsBDaily['date'] \
<= x['date']].agg({'date': 'max', 'visitorsPerDateB': 'sum'}), axis=1)

visitorsBCummulative.columns = ['date', 'visitorsCummulativeB']

ordersADaily = orders[orders['group'] == 'A'][['date', 'transaction_id', 'visitor_id', 'revenue']].groupby('date', \
as_index=False).agg({'transaction_id': pd.Series.nunique, 'revenue': 'sum'})

ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']

ordersACummulative = ordersADaily.apply(lambda x: ordersADaily[ordersADaily['date'] \
<= x['date']].agg({'date': 'max', 'ordersPerDateA': 'sum', 'revenuePerDateA': 'sum'}), axis=1).sort_values(by=['date'])

ordersACummulative.columns = ['date', 'ordersCummulativeA', 'revenueCummulativeA']

ordersBDaily = orders[orders['group'] == 'B'][['date', 'transaction_id', 'visitor_id', 'revenue']].groupby('date', \
as_index=False).agg({'transaction_id': pd.Series.nunique, 'revenue': 'sum'})

ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']

ordersBCummulative = ordersBDaily.apply(lambda x: ordersBDaily[ordersBDaily['date'] \
<= x['date']].agg({'date': 'max', 'ordersPerDateB': 'sum', 'revenuePerDateB': 'sum'}), axis=1).sort_values(by=['date'])

ordersBCummulative.columns = ['date', 'ordersCummulativeB', 'revenueCummulativeB']

data = (
    ordersADaily.merge(
        ordersBDaily, left_on='date', right_on='date', how='left'
    )
    .merge(ordersACummulative, left_on='date', right_on='date', how='left')
    .merge(ordersBCummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsADaily, left_on='date', right_on='date', how='left')
    .merge(visitorsBDaily, left_on='date', right_on='date', how='left')
    .merge(visitorsACummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsBCummulative, left_on='date', right_on='date', how='left')
)

data.head()
```

Создадим переменные, содержащие данные о пользователях, совершивших как минимум один заказ - id пользователя и количество его заказов

```python
ordersByUsersA = orders[orders['group'] == 'A'].groupby('visitor_id', as_index=False).agg({'transaction_id': pd.Series.nunique})

ordersByUsersA.columns = ['visitor_id', 'orders']

ordersByUsersB = orders[orders['group'] == 'B'].groupby('visitor_id', as_index=False).agg({'transaction_id': pd.Series.nunique})

ordersByUsersB.columns = ['visitor_id', 'orders'] 
```

Создадим переменные с количеством заказов для каждого пользователя, в том числе и для тех, кто не совершил ни одного заказа.

```python
sampleA = pd.concat([ordersByUsersA['orders'], pd.Series(0, index=np.arange(data['visitorsPerDateA'].sum() \
- len(ordersByUsersA['orders'])), name='orders')], axis=0)

sampleB = pd.concat([ordersByUsersB['orders'], pd.Series(0, index=np.arange(data['visitorsPerDateB'].sum() \
- len(ordersByUsersB['orders'])), name='orders')], axis=0) 
```

Посчитаем статистическую значимость различий в среднем количестве заказов между группами по «сырым» данным.

Для этого применим критерий Манна-Уитни и отформатируем p-value, округлив его до трёх знаков после запятой.

```python
print("P-value = ""{0:.3f}".format(stats.mannwhitneyu(sampleA, sampleB)[1]))
print("Относительное различие = ""{0:.3f}".format(sampleB.mean() / sampleA.mean() - 1))
```

**Вывод**: P-value меньше значения 0.05. Значит, нулевую гипотезу отвергаем в пользу альтернативной и считаем, что различия в среднем количестве заказов есть.

Относительный выигрыш группы B равен 13,8%.

### Статистическая значимость различий в среднем чеке заказа между группами по «сырым» данным

Сформулируем гипотезы. 

**Нулевая**: различий в среднем чеке заказа между группами нет. 

**Альтернативная**: различия в среднем чеке заказа между группами есть.

```python
print("P-value = "'{0:.3f}'.format(stats.mannwhitneyu(orders[orders['group']=='A']['revenue'], \
                                                      orders[orders['group']=='B']['revenue'])[1]))
print("Относительное различие = "'{0:.3f}'.format(orders[orders['group']=='B']['revenue'].mean()/\
                                                  orders[orders['group']=='A']['revenue'].mean()-1)) 
```

**Вывод**: P-value значительно выше значения 0.05. Значит, причин отвергать нулевую гипотезу и считать, что в среднем чеке есть различия, нет. 

Однако видим, что средний чек группы B выше среднего чека группы A на 25,9%. Исходя из этого, наблюдаемые различия не являются значимыми при таком уровне значимости, а следовательно были получены случайно.

Узнаем, как их действия повлияли на результаты теста. Сначала подготовим выборки количества заказов по пользователям по группам теста.

```python
sampleAFiltered = pd.concat([ordersByUsersA[np.logical_not(ordersByUsersA['visitor_id'].isin(abnormalUsers))]['orders'], \
pd.Series(0, index=np.arange(data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])), name='orders')], axis=0)

sampleBFiltered = pd.concat([ordersByUsersB[np.logical_not(ordersByUsersB['visitor_id'].isin(abnormalUsers))]['orders'], \
pd.Series(0, index=np.arange(data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])),name='orders')], axis=0) 
```

Гипотезы аналогичны предыдущим. 

**Нулевая**: различий в среднем количестве заказов между группами нет. 

**Альтернативная**: различия в среднем количестве заказов между группами есть.

```python
print("P-value = "'{0:.3f}'.format(stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))
print("Относительное различие = "'{0:.3f}'.format(sampleBFiltered.mean()/sampleAFiltered.mean()-1)) 
```

**Вывод**: P-value меньше значения 0.05. Значит, нулевую гипотезу отвергаем в пользу альтернативной и считаем, что различия в среднем количестве заказов есть.

Относительный выигрыш группы B равен 17,3%.

Результат по "очищенным" данным подтверждает выводы, сделанные по "сырым" данным.

### Статистическая значимость различий в среднем чеке заказа между группами по «очищенным» данным

Гипотезы:

**Альтернативная**: различия в среднем чеке между группами есть.

```python
print("P-value = "'{0:.3f}'.format(stats.mannwhitneyu(orders[np.logical_and(orders['group'] == 'A',\
np.logical_not(orders['visitor_id'].isin(abnormalUsers)))]['revenue'],
orders[np.logical_and(orders['group'] == 'B',\
np.logical_not(orders['visitor_id'].isin(abnormalUsers)))]['revenue'])[1]))

print("Относительное различие = ""{0:.3f}".format(orders[np.logical_and(orders['group'] == 'B', \
np.logical_not(orders['visitor_id'].isin(abnormalUsers)))]['revenue'].mean() \
/ orders[np.logical_and(orders['group'] == 'A', \
np.logical_not(orders['visitor_id'].isin(abnormalUsers)))]['revenue'].mean()- 1)) 
```

При этом, средний чек группы B - ниже среднего чека группы A на 2%. То есть, все преимущество группы B задавали аномалии, а на самом деле фактически разницы между двумя группами нет (группа B даже чуть хуже).
